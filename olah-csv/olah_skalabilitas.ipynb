## Code Untuk Olah Dataset Skalabilitas

### Dependencies

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dotenv import load_dotenv
```

### Load ENV Variables

```python
load_dotenv(override=True)
```

### Masukkan Lokasi File

```python
location_platform_1 = os.environ.get('LOCATION_PLATFORM_1')
location_platform_2 = os.environ.get('LOCATION_PLATFORM_2')
location_platform_3 = os.environ.get('LOCATION_PLATFORM_3')
```

### Membaca File CSV

```python
# Platform 1
data_platform_1 = pd.read_csv(location_platform_1)
data_1vm_platform_1 = data_platform_1[data_platform_1['vm_size'] == 1]
data_2vm_platform_1 = data_platform_1[data_platform_1['vm_size'] == 2]
data_3vm_platform_1 = data_platform_1[data_platform_1['vm_size'] == 3]

data_platform_2 = pd.read_csv(location_platform_2)
data_1vm_platform_2 = data_platform_2[data_platform_2['vm_size'] == 1]
data_2vm_platform_2 = data_platform_2[data_platform_2['vm_size'] == 2]
data_3vm_platform_2 = data_platform_2[data_platform_2['vm_size'] == 3]

data_platform_3 = pd.read_csv(location_platform_3)
data_1vm_platform_3 = data_platform_3[data_platform_3['vm_size'] == 1]
data_2vm_platform_3 = data_platform_3[data_platform_3['vm_size'] == 2]
data_3vm_platform_3 = data_platform_3[data_platform_3['vm_size'] == 3]
```
